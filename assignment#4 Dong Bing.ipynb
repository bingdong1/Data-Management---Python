```python
#Create data frame for selected column
accident2014_select2 = accident2014[['c5','c6','c143','c15','c76','c78','c77']]
```

```python
#Rename the active data frame column in order to combine the dataset together
accident2014_select2=accident2014_select2.rename(columns={'c5':'UNIQUE_ID'}) 
accident2014_select2=accident2014_select2.rename(columns={'c6':'YEAR'}) 
accident2014_select2=accident2014_select2.rename(columns={'c143':'AIRPORT_ID'}) 
accident2014_select2=accident2014_select2.rename(columns={'c15':'AIRPORT_NAME'}) 
accident2014_select2=accident2014_select2.rename(columns={'c76':'FATALITY'}) 
accident2014_select2=accident2014_select2.rename(columns={'c78':'PRIMARY_CAUSE_FACTOR_CODE'})
accident2014_select2=accident2014_select2.rename(columns={'c77':'PRIMARY_CAUSE_FACTOR_TEXT'}) 
```

```python
#Rename the value to combine all the airport id, defintion obtain from ATA numering system
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '85'] = 'RECIPROCATING ENGINE'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '73'] = 'ENGINE FUEL AND CONTROL'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '57'] = 'WINGS'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '61'] = 'PROPELLERS/PROPULSION'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '72'] = 'ENGINE TURBINE/TURBO PROP DUCTED FAN/UNDUCTED FAN'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '27'] = 'FLIGHT CONTROLS'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '52'] = 'DOORS'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '71'] = 'POWER PLANT'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '67'] = 'ROTORS FLIGHT CONTROL'
accident2014_select2_filter.c78[accident2014_select2_filter.c78 == '28'] = 'FUEL'
```

```python
def set_color(row):
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "85":
        return "RECIPROCATING ENGINE"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "73":
        return "ENGINE FUEL AND CONTROL"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "57":
        return "WINGS"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "61":
        return "PROPELLERS/PROPULSION"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "72":
        return "ENGINE TURBINE/TURBO PROP DUCTED FAN/UNDUCTED FAN"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "27":
        return "FLIGHT CONTROLS"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "52":
        return "DOORS"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "71":
        return "POWER PLANT"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "67":
        return "ROTORS FLIGHT CONTROL"
    if row["PRIMARY_CAUSE_FACTOR_CODE"] == "28":
        return "FUEL"
    else:
        return "ALL OTHER"
```

```python
accident2014_select2_filter = accident2014_select2_filter.assign(defintion=accident2014_select2_filter.apply(set_color, axis=1))
```

```python
accident_df = accident2014_select2_filter.defintion.value_counts().head(11)
```
